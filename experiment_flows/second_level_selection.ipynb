```python
from model_pipelines.first_level_feature_selectors import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=pd.read_csv("C:\\Users\\andrz\\Downloads\\data\\data\\y_train.txt")
x=pd.read_csv("C:\\Users\\andrz\\Downloads\\data\\data\\x_train.txt",sep=" ")
#select only the features marked by boruta previously
boruta_features=['V5', 'V11', 'V32', 'V46', 'V60', 'V117', 'V120', 'V160', 'V161',
       'V175', 'V176', 'V191', 'V199', 'V215', 'V224', 'V227', 'V255', 'V262',
       'V265', 'V309', 'V313', 'V329', 'V342', 'V345', 'V380', 'V390', 'V416',
       'V443', 'V483']
x=x.loc[:,boruta_features]
```

```python
selected_features_dictionary={
    'rf_1000_1':{},
    'rf_1000_3':{},
    'rf_3000_1':{},
    'rf_3000_3':{},
    'xgb_1000_1':{},
    'xgb_1000_3':{},
    'xgb_3000_1':{},
    'xgb_3000_3':{},
    'shap_rf_1000_1':{},
    'shap_rf_1000_3':{},
    'shap_rf_3000_1':{},
    'shap_rf_3000_3':{},
    'shap_xgb_1000_1':{},
    'shap_xgb_1000_3':{},
    'shap_xgb_3000_1':{},
    'shap_xgb_3000_3':{}
}
```

```python
#one-run selection, does not depend on random state
selected_features_dictionary['KS']=Kolmogorov_Smirnov_selector(x,y,12)[0]
selected_features_dictionary['Corr']=select_best_correlation(x,y,12)
selected_features_dictionary['MI']=select_best_mutual_information(x,y,12)
```

```python
random_seeds = [42, 2137, 2026]
N_FEATURES = 12
for seed in random_seeds:
    print(f"\n=========================================")
    print(f"RUNNING ALL SELECTORS FOR SEED: {seed}")
    print(f"=========================================\n")
    # --- 1. Random Forest Selectors ---
    selected_features_dictionary['rf_1000_1'][seed] = random_forest_selector(
        x, y, n_features=N_FEATURES, n_estimators=1000, max_depth=1, random_state=seed
    )
    selected_features_dictionary['rf_1000_3'][seed] = random_forest_selector(
        x, y, n_features=N_FEATURES, n_estimators=1000, max_depth=3, random_state=seed
    )
    selected_features_dictionary['rf_3000_1'][seed] = random_forest_selector(
        x, y, n_features=N_FEATURES, n_estimators=3000, max_depth=1, random_state=seed
    )
    selected_features_dictionary['rf_3000_3'][seed] = random_forest_selector(
        x, y, n_features=N_FEATURES, n_estimators=3000, max_depth=3, random_state=seed
    )
    # --- 2. XGBoost Selectors ---
    selected_features_dictionary['xgb_1000_1'][seed] = xgb_selector(
        x, y, n_features=N_FEATURES, n_estimators=1000, max_depth=1, random_state=seed, return_='list'
    )
    selected_features_dictionary['xgb_1000_3'][seed] = xgb_selector(
        x, y, n_features=N_FEATURES, n_estimators=1000, max_depth=3, random_state=seed, return_='list'
    )
    selected_features_dictionary['xgb_3000_1'][seed] = xgb_selector(
        x, y, n_features=N_FEATURES, n_estimators=3000, max_depth=1, random_state=seed, return_='list'
    )
    selected_features_dictionary['xgb_3000_3'][seed] = xgb_selector(
        x, y, n_features=N_FEATURES, n_estimators=3000, max_depth=3, random_state=seed, return_='list'
    )

    # --- 3. SHAP Selectors (with plot presentation turned off) ---
    selected_features_dictionary['shap_rf_1000_1'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='rf', n_estimators=1000, max_depth=1, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_rf_1000_3'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='rf', n_estimators=1000, max_depth=3, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_rf_3000_1'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='rf', n_estimators=3000, max_depth=1, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_rf_3000_3'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='rf', n_estimators=3000, max_depth=3, present_shap_plot=False, random_state=seed
    )

    selected_features_dictionary['shap_xgb_1000_1'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='xgb', n_estimators=1000, max_depth=1, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_xgb_1000_3'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='xgb', n_estimators=1000, max_depth=3, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_xgb_3000_1'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='xgb', n_estimators=3000, max_depth=1, present_shap_plot=False, random_state=seed
    )
    selected_features_dictionary['shap_xgb_3000_3'][seed] = SHAP_selector(
        x, y, n_features=N_FEATURES, modeltype='xgb', n_estimators=3000, max_depth=3, present_shap_plot=False, random_state=seed
    )
```
